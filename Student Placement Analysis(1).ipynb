```python
# To check the missing values
df.isnull().sum()
```

#  Grouping & Visualisation:

```python
#To check how many students are present in each branch
df['Branch'].value_counts()
```

```python
# Visualizing the data using seaborn and matplotlib libraries
sns.countplot(y='Branch', data=df)
plt.title("Students by Branch")
plt.show()
```

```python
# To gain information on which companies students got high placements
df['Company'].value_counts().head(10)
```

```python
df['Company'].value_counts().head(10).plot(kind='bar')
plt.title("Top Recruiting Companies")
plt.show()
```

```python
# To display the number of job roles
df['Job Role'].value_counts().head(10)
```

```python
sns.countplot(y='Job Role', data=df)
plt.show()
```

```python
#To display only 'Web Developer' job role records
print(df[df['Job Role'] == 'Web Developer'])
```

```python
sns.histplot(df['Salary (INR)'], kde=True)
plt.title("Salary Distribution")
plt.show()
```

```python
# To display the student details who got highest salary
df.loc[df['Salary (INR)'].idxmax()]
```

```python
#To display the average salary with respect to the each branch
avg_salary = df.groupby('Branch')['Salary (INR)'].mean()
avg_salary
```

```python
avg_salary.plot(kind='bar')
plt.title("Average Salary by Branch")
plt.show()
```

```python
#To display the average salary with respect to the each company
avg_salary = df.groupby('Company')['Salary (INR)'].mean()
avg_salary
```

```python
avg_salary.plot(kind='bar')
plt.title("Average Salary by Company")
plt.show()
```

```python
# pie chart for top 5 companies
df['Company'].value_counts().head(5).plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)

plt.title("Top 5 Recruiting Companies")
plt.ylabel("")
plt.show()
```

```python
# To display the how many students got placements with respect to the locations
df['Location'].value_counts().head(10)
```

```python
df['Location'].value_counts().head(10).plot(kind='bar')
plt.title("Top Placement Locations")
plt.show()
```

```python
#To display how many students are completed their graduation in every year 
df['Graduation Year'].value_counts().sort_index()
```

```python
sns.countplot(x='Graduation Year', data=df)
plt.show()
```

```python
#To display how many students are got the placements in every year 
df['Placement Date'] = pd.to_datetime(df['Placement Date'])
df['Placement Date'].dt.year.value_counts().sort_index()
```
