```python
#Loan Approval System – Real-Time Banking Task
Applicant_Name=input("Enter apllicant name: ")
Age=int(input("Enter apllicant age: "))
Monthly_Salary=int(input("Enter apllicant salary:"))
Credit_Score=int(input("Enter apllicant credit score: "))
Number_of_Existing_Loans=int(input("Enter number of apllicant existing loans: "))
Employment_Status=input("Enter apllicant employment status(employed / unemployed): ")
Loan_Amount_Required=int(input("Enter amount of money required: "))
if Age>=21 and Age<=60:
    if Monthly_Salary>40000:
        if Credit_Score>750 or Credit_Score==750:
            if Number_of_Existing_Loans<2:
                if Employment_Status=="employed":
                    if Loan_Amount_Required<10*Monthly_Salary:
                             print("Loan Approved")
                    else:
                        print(f"Loan Rejected because Loan amount exceeds eligibility{Loan_Amount_Required} ")
                else:
                    print(f"Loan Rejected because Applicant is unemployed{Employment_Status}")
            else:
                print(f"Loan Rejected because Applicant has Too many existing loansNumber of Existing Loans{Number_of_Existing_Loans}")
        else:
            print(f"Loan Rejected because Applicant has poor credit score{Credit_Score}")
    else:
        print(f"Loan Rejected because salary is too low{Monthly_salary}")
else:
    print(f"Loan Rejected because Age criteria not satisfied{Age}")
```
