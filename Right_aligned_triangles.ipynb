```python
# Task 1: Right-Aligned Increasing Triangle
rows=5
for outer_loop in range(1,rows+1):
  for space in range(rows-outer_loop):
    print(" ",end="")
  for inner_loop in range(1,outer_loop+1):
    print("*",end="")
  print()
```

```python
# Task 2: Right-Aligned Decreasing Triangle
rows=5
for outer_loop in range(1,rows+1):
  for space in range(outer_loop-1):
    print(" ",end="")
  for inner_loop in range(rows-outer_loop+1):
    print("*",end="")
  print()
```
