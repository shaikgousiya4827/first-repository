```python
#Task: Smart Password Validator
password=input("Checking password:")
print("First 3 chars:",password[:3])
print("Last 3 chars:",password[-3:])
lower=0
upper=0
digit=0
spl_character=0
for i in password:
    if i.islower():
        lower+=1
    elif i.isupper():
        upper+=1
    elif i.isdigit():
        digit+=1
    else:
        spl_character+=1
length_check=False
i=0
while i < len(password):
  if len(password)>=8:
    length_check=True
    break
  i+=1
if length_check==True and lower>=1 and upper>=1 and digit>=1 and spl_character>=1:
  print("Strong Password")
else:
  print("Weak Password")
if len(password)>=8:
  masked=password[0]+"*"*len(password[1:-1])+password[-1]
  print(masked)
digit=0
for i in password:
  if i.isdigit():
        digit+=1
print(digit)
upper=0
for i in password:
  if i.isupper():
    upper+=1
print(upper)
a=len(password)//2
print("reverse first half of password is", password[0:a:][::-1]+password[a::])
if password[0].isupper() and password[-1].isdigit():
  print("starts with uppercase and ends with digit")
else:
  print("does not start with uppercase and end with digit")

```
