```python
##Task 1: ATM Withdrawal System
account_balance=int(input("Enter your account balance"))
withdrawal_amount=int(input("Enter withdrawal amount"))
try:
  if withdrawal_amount<=0 or withdrawal_amount>account_balance:
    raise ValueError
  else:
   remaining_balance=account_balance-withdrawal_amount
   print(f"remaining balance :",remaining_balance)
except ValueError:
  print("invalid withdrawal amount")
finally:
  print("Thank you for using our ATM.")

```

```python
##Task 2: Student Result Processing System (Custom Exception)
class FailException(Exception):
  pass
marks=int(input("Enter marks of one subject"))
try:
    if marks<0 or marks>100:
      raise ValueError
    elif marks<35:
      raise FailException
except ValueError:
    print("Invalid marks")
except FailException:
    print("Student Failed")
else:
    print("Student Passed")
finally:
    print("Result Processing Completed")




```
