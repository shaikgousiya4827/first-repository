```python
#Task 1: Student Attendance Report Generator
file=open("attendance.txt","r")
multiple_lines=file.read()
count_p=0
count_a=0
for i in multiple_lines.split("\n"):
  a=i.split(",")[1]
  if a=="Present":
    count_p+=1
  else:
    count_a+=1
file=open("report.txt","w+")
file.write("Attendance Report\n")
file.write("-----------------")
file.write("\nTotal Students :"+str(count_p+count_a))
file.write("\nPresent Students :"+str(count_p))
file.write("\nAbsent Students :"+str(count_a))
print("Report generated successfully in report.txt")
file.close()
```

```python
#Task 2: Log File Cleaner (Interview-Level)
file=open("logs.txt","w+")
file.write("INFO: User logged in\n")
file.write("ERROR: Database connection failed\n")
file.write("INFO: File uploaded\n")
file.write("ERROR: Invalid password\n")
file.write("INFO: User logged out\n")
file=open("logs.txt","r")
multiple_lines=file.read()
count=0
file1=open("errors.txt","w+")
for i in multiple_lines.split("\n"):
  a=i.split(":")[0]
  if a=="ERROR":
    file1.write(i + "\n")
    count+=1
print("Total Errors Found:",count)
file.close()
file1.close()
```
