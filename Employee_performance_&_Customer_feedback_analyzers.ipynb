```python
#Task1 :Employee Performance Analyzer
emp_names="Satya,Ravi,Priya,Kiran"
ratings=90,75,95,80
a=list(emp_names.split(","))
b=list(ratings)
print(f"Employee with Highest Rating : {a[b.index(max(b))]} ({max(b)})")
print(f"Employee with Lowest Rating : {a[b.index(min(b))]} ({min(b)})")
c=(max(b)+min(b))/2
print("average Rating :",c)
print("Above Average Employees :")
for i in b:
  if i>c:
    print(a[b.index(i)])
```

```python
input="Python learning is interesting"
def hi():
  s=input.split(" ")
  print("total Words:",len(s))
hi()
def vowels():
  a="aeiou"
  count=0
  for i in input:
    if i in a:
     count+=1
  print("Total Vowels:",count)
vowels()
def long():
  s=input.split(" ")    ##we also replace this long() function code with this **print("Longest word:",max(s,key=len))**
  a=[]
  b=[]
  for i in s:
    a.append(i)
    b.append(len(i))
  print("Longest Word :",a[b.index(max(b))])
long()
def reverse():
  s=input.split(" ")
  for i in s:
    print(i[::-1],end=" ")
reverse()
def frequency():
  print("\nCharacter frequency:")
  input1=input.replace(" ","")
  count=0
  for i in input1:
    print(i,":",input.count(i))
frequency()

```
