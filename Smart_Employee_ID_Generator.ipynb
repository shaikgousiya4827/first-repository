```python
# 1️⃣ Smart Employee ID Generator 🆔(By using slicing)
Employee_Name=input("Enter Employee Name:")
Joining_Year=int(input("Enter Joining Year:"))
Department_Name=input("Enter Department Name:")

Employee_ID = Employee_Name[0:3:1].upper()+str(Joining_Year)[-2:].upper()+Department_Name[0:3:1].upper()

print("Generated Employee ID :", Employee_ID)
```

```python
# 1️⃣ Smart Employee ID Generator 🆔(By using indexing)
Employee_Name=input("Enter Employee Name:")
Joining_Year=int(input("Enter Joining Year:"))
Department_Name=input("Enter Department Name:")

Employee_ID = (Employee_Name[0]+Employee_Name[1]+Employee_Name[2]).upper()+(str(Joining_Year)[-2]+str(Joining_Year)[-1]).upper()+(Department_Name[0]+Department_Name[1]+Department_Name[2]).upper()

print("Generated Employee ID :", Employee_ID)
```
